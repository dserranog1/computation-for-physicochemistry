```python
import numpy as np
import matplotlib.pyplot as plt

def random_walk_1d(num_steps=100, num_walks=5):
    """
    Simulates multiple 1D random walks and plots them.
    
    Parameters:
    num_steps (int): Number of steps for each random walk.
    num_walks (int): Number of random walks to simulate.
    """
    plt.figure(figsize=(10, 6))
    
    for _ in range(num_walks):
        position = np.zeros(num_steps + 1)
        
        for step in range(1, num_steps + 1):
            random_number = np.random.rand()  # Random number between 0 and 1
            step_direction = 1 if random_number >= 0.48 else -1  # +1 or -1 step
            position[step] = position[step - 1] + step_direction  # Update position
        
        # Plot this random walk
        plt.plot(range(num_steps + 1), position, linestyle="-", alpha=0.7)
    
    plt.title(f"{num_walks} Random Walks in 1D")
    plt.xlabel("Step #")
    plt.ylabel("Position")
    plt.grid(True)
    plt.show()

random_walk_1d(num_steps=10000, num_walks=10)
```
